```python
import pandas as pd
df = pd.read_csv('data/complete_membership_payment_history_v1.csv')
df.head(60)
```

```python
# Calculate the percentage of rows with a value of 0 for the 'Monto' column for each year, quarter, and month
percent_zero_year = df[df['Monto'] == 0].groupby('year').size() / df.groupby('year').size() * 100
percent_zero_quarter = df[df['Monto'] == 0].groupby(['year', 'quarter']).size() / df.groupby(['year', 'quarter']).size() * 100
percent_zero_month = df[df['Monto'] == 0].groupby(['year', 'month']).size() / df.groupby(['year', 'month']).size() * 100

# Print the results
print('Percentage of rows with a value of 0 for the "Monto" column for each year:')
print(percent_zero_year.round(2).apply('{:.2f}%'.format))
print('\n\nPercentage of rows with a value of 0 for the "Monto" column for each quarter:')
print(percent_zero_quarter.round(2).apply('{:.2f}%'.format))
```

```python
import pandas as pd

def fill_zeros(df):
    # create a copy of the dataframe
    df_filled = df.copy()
    
    # iterate through the rows of the dataframe
    for index, row in df.iterrows():
        # if the value in the 'Monto' column is 0
        if row['Monto'] == 0:
            # find the first non-zero value in the 'Monto' column for the same month, quarter, year, and plan
            # and store it in the variable 'value'
            non_zero_rows = df[(df['month'] == row['month']) & (df['quarter'] == row['quarter']) & 
                       (df['year'] == row['year']) & (df['Plan'] == row['Plan']) & (df['Monto'] != 0)]
            if non_zero_rows.empty:
                # if no non-zero rows were found, set 'value' to 0
                value = 0
            else:
                # if non-zero rows were found, set 'value' to the first one
                value = non_zero_rows['Monto'].iloc[0]
            # set the value of the 'Monto' column in the copy dataframe to 'value'
            df_filled.at[index, 'Monto'] = value
    return df_filled
```

```python
df = fill_zeros(df)
```
